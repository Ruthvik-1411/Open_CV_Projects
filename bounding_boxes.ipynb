```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
im1=cv2.imread('images/faces/Foundation_Photo_010201.jpg')
im1=cv2.cvtColor(im1,cv2.COLOR_BGR2RGB)
im2=im1.copy()
im3=im1.copy()
```

```python
fig=plt.figure(figsize=(30,30))

fig.add_subplot(1,3,1)
plt.imshow(im1)
fig.add_subplot(1,3,2)
plt.imshow(im2)
fig.add_subplot(1,3,3)
plt.imshow(im3)
```

```python
text='ADMIN'
color=(255,255,0)
```

# ADMIN Bounded Marking

```python
def mark_face_v2(img,x,y,w,h):
    factor=round((1.0*h)/(9*22),1)
    for each in range(0,10,2):
        cv2.line(img,(x+(each*(round(w/9))),y),((x+((each+1)*(round(w/9)))),y),color,5)
        cv2.line(img,(x+(each*(round(w/9))),y+h),((x+((each+1)*(round(w/9)))),y+h),color,5)
        cv2.line(img,(x,y+(each*round(h/9))),(x,y+((each+1)*(round(h/9)))),color,5)
        cv2.line(img,(x+w,y+(each*round(h/9))),(x+w,y+((each+1)*(round(h/9)))),color,5)
    cv2.rectangle(img,(x+round((2*w/9)),y+h+round((0.2*h/9))),(x+w-round((2*w/9)),y+h+round((1.5*h/9))),color,-1)
    cv2.putText(img,text,(x+round((2.5*w/9)),y+h+round((1.2*h/9))),cv2.FONT_HERSHEY_DUPLEX,color=(0,0,0),fontScale=factor,thickness=2)
```

# Rectangle Crosshair 'Target'

```python
def rmark(img,x,y,w,h):
    cv2.line(img,(x,y),(x+round(w/5),y),(255,255,0),7)
    cv2.line(img,(x+w-round(w/5),y),(x+w,y),(255,255,0),7)
    cv2.line(img,(x,y+h),(x+round(w/5),y+h),(255,255,0),7)
    cv2.line(img,(x+w-round(w/5),y+h),(x+w,y+h),(255,255,0),7)
    ##########
    cv2.line(img,(x,y),(x,y+round(h/5)),(255,255,0),7)
    cv2.line(img,(x,y+h-round(h/5)),(x,y+h),(255,255,0),7)
    cv2.line(img,(x+w,y),(x+w,y+round(h/5)),(255,255,0),7)
    cv2.line(img,(x+w,y+h-round(h/5)),(x+w,y+h),(255,255,0),7)
    ##########
    cv2.circle(img,(x+round(w/2),y),3,(255,255,0),-1)
    cv2.circle(img,(x+round(w/2),y+h),3,(255,255,0),-1)
    cv2.circle(img,(x,y+round(h/2)),3,(255,255,0),-1)
    cv2.circle(img,(x+w,y+round(h/2)),3,(255,255,0),-1)
    ######
    cv2.circle(img,(x+round(w/2),y+round(h/2)),3,(255,255,0),-1)
    cv2.rectangle(img,(x-2*round(w/2),y-2*round(h/2)),(x+w+2*round(w/2),y+h+2*round(h/2)),(255,255,0),7)
    ########
    cv2.line(img,(x+round(w/2),y-2*round(h/2)),(x+round(w/2),y-round(h/2)),(255,255,0),7)
    cv2.line(img,(x+round(w/2),y+h+2*round(h/2)),(x+round(w/2),y+h+round(h/2)),(255,255,0),7)
    cv2.line(img,(x-2*round(w/2),y+round(h/2)),(x-round(w/2),y+round(h/2)),(255,255,0),7)
    cv2.line(img,(x+w+2*round(w/2),y+round(h/2)),(x+w+round(w/2),y+round(h/2)),(255,255,0),7)
```

# Circular Crosshair 'Target'

```python
def omark(img,x,y,w,h):
    cv2.line(img,(x,y),(x+round(w/5),y),(255,255,0),7)
    cv2.line(img,(x+w-round(w/5),y),(x+w,y),(255,255,0),7)
    cv2.line(img,(x,y+h),(x+round(w/5),y+h),(255,255,0),7)
    cv2.line(img,(x+w-round(w/5),y+h),(x+w,y+h),(255,255,0),7)
    ##########
    cv2.line(img,(x,y),(x,y+round(h/5)),(255,255,0),7)
    cv2.line(img,(x,y+h-round(h/5)),(x,y+h),(255,255,0),7)
    cv2.line(img,(x+w,y),(x+w,y+round(h/5)),(255,255,0),7)
    cv2.line(img,(x+w,y+h-round(h/5)),(x+w,y+h),(255,255,0),7)
    ##########
    cv2.circle(img,(x+round(w/2),y),3,(255,255,0),-1)
    cv2.circle(img,(x+round(w/2),y+h),3,(255,255,0),-1)
    cv2.circle(img,(x,y+round(h/2)),3,(255,255,0),-1)
    cv2.circle(img,(x+w,y+round(h/2)),3,(255,255,0),-1)
    ######
    cv2.circle(img,(x+round(w/2),y+round(h/2)),3,(255,255,0),-1)
    cv2.circle(img,(x+round(w/2),y+round(h/2)),round(1.5*(round((w+h)/2))),(255,255,0),4)
    ########
    cv2.line(img,(x+round(w/2),y-2*round(h/2)),(x+round(w/2),y-round(h/2)),(255,255,0),7)
    cv2.line(img,(x+round(w/2),y+h+2*round(h/2)),(x+round(w/2),y+h+round(h/2)),(255,255,0),7)
    cv2.line(img,(x-2*round(w/2),y+round(h/2)),(x-round(w/2),y+round(h/2)),(255,255,0),7)
    cv2.line(img,(x+w+2*round(w/2),y+round(h/2)),(x+w+round(w/2),y+round(h/2)),(255,255,0),7)
```

```python
haar=cv2.CascadeClassifier('Open cv/haarcascades/haarcascade_frontalface_default.xml')
```

```python
gray_im1=cv2.cvtColor(im1,cv2.COLOR_BGR2GRAY)
faces=haar.detectMultiScale(gray_im1,1.31)
for (x,y,w,h) in faces:
    mark_face_v2(im1,x,y,w,h)
```

```python
gray_im2=cv2.cvtColor(im2,cv2.COLOR_BGR2GRAY)
faces=haar.detectMultiScale(gray_im2,1.31)
for (x,y,w,h) in faces:
    rmark(im2,x,y,w,h)
```

```python
gray_im3=cv2.cvtColor(im3,cv2.COLOR_BGR2GRAY)
faces=haar.detectMultiScale(gray_im3,1.31)
for (x,y,w,h) in faces:
    omark(im3,x,y,w,h)
```

```python
fig=plt.figure(figsize=(30,30))

fig.add_subplot(1,3,1)
plt.imshow(im1)
fig.add_subplot(1,3,2)
plt.imshow(im2)
fig.add_subplot(1,3,3)
plt.imshow(im3)
```
